```python
#IMporting libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset= load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head()
```

```python
data_science_jobs = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Scientist')]
data_science_jobs = data_science_jobs[['job_title', 'job_skills', 'job_posted_date']]
data_science_jobs = data_science_jobs[:10]
data_science_jobs = data_science_jobs.reset_index(drop=True)
data_science_jobs
```

```python
data_science_jobs_list = data_science_jobs.to_dict(orient='records')
data_science_jobs_list
```

```python
from datetime import datetime

# Convert all job_posted_date entries to strings
for job in data_science_jobs_list:
    job['job_posted_date'] = job['job_posted_date'].strftime('%Y-%m-%d')
    job['job_posted_date'] = datetime.strptime(job['job_posted_date'], '%Y-%m-%d')



data_science_jobs_list
```

```python
import ast

for job in data_science_jobs_list:
    if isinstance(job['job_skills'], str):
        job['job_skills'] = ast.literal_eval(job['job_skills'])
data_science_jobs_list
```
