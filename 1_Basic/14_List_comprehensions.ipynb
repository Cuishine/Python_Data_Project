# List Comprehensions

## Notes

* A way way to create a new list (with shorter syntax) based on the values of an existing list.

Not limited to only `list` comprehension: 
- `set` comprehension
- `tuple` comprehension
- `dictionary` comprehension

## Importance

Provide a concise way to create lists. Useful for data manipulation and filtering in pandas.

```python
numbers = [x for x in range(100)]

```

```python
numbers = []

for number in range(100):
    numbers.append(number)
```

```python
numbers = {x for x in 'PYTHON'}
numbers
```

```python
#IMporting libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


# Loading Data
dataset= load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_title= list(df['job_title'])

len(job_title)
```

```python
job_title_analyst = []

for job in job_title:
    if job is not None and 'Data Analyst' in job:
        job_title_analyst.append(job)
```

```python
analyst_list = [job for job in job_title if job is not None and 'Data Analyst' in job]
len(analyst_list)
```

```python
print("Original list is:", len(job_title), "jobs")
print("New list is: ", len(analyst_list), "jobs")
```
